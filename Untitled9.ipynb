```python
from sklearn.metrics import confusion_matrix, mean_squared_error
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_predict

#
# These functions are used for evaluation regression models
#

def display_predict_rmse(reg, X, y, label='RMSE'):
    """Display RMSE of model predictions"""
    y_pred = reg.predict(X)
    rmse = np.sqrt(mean_squared_error(y, y_pred))
    print(f'{label}: {rmse:.4f}')

def display_cv_predict_rmse(reg, X, y, label='Cross Validation RMSE'):
    """Display RMSE of predictions from cross-validation"""
    y_pred = cross_val_predict(reg, X, y, cv=5, n_jobs=-1)
    print(f'{label}: {np.sqrt(mean_squared_error(y, y_pred)):.4f}')

#
# These functions is used for evaluating classification models
#

def display_confusion_matrix_and_scores(clf, X, y):
    """Display confusion matrix and various scores"""
    y_pred = clf.predict(X)
    print('Confusion Matrix     ',
          f'Precision = {precision_score(y, y_pred):6.4f}, ',
          f'Recall = {recall_score(y, y_pred):6.4f}, ',
          f'F1 = {f1_score(y, y_pred):6.4f}, ',
          f'Accuracy = {accuracy_score(y, y_pred):6.4f}')
    cm = confusion_matrix(y, y_pred)
    print(f'[ {cm[0,0]:4d} {cm[0,1]:4d} ]')
    print(f'[ {cm[1,0]:4d} {cm[1,1]:4d} ]')

def display_cv_confusion_matrix_and_scores(clf, X, y):
    """Display confusion matrix and various scores from cross-validation"""
    y_pred = cross_val_predict(clf, X, y, cv=5, n_jobs=-1)
    print('Confusion Matrix     ',
          f'Precision = {precision_score(y, y_pred):6.4f}, ',
          f'Recall = {recall_score(y, y_pred):6.4f}, ',
          f'F1 = {f1_score(y, y_pred):6.4f}, ',
          f'Accuracy = {accuracy_score(y, y_pred):6.4f}')
    cm = confusion_matrix(y, y_pred)
    print(f'[ {cm[0,0]:4d} {cm[0,1]:4d} ]')
    print(f'[ {cm[1,0]:4d} {cm[1,1]:4d} ]')
```

Indicators like SMA (Simple Moving Average) and Rolling Standard Deviation are tools that help analyze patterns and trends in data. SMA shows the average price over a certain number of days, like 20 or 50, to smooth out short-term fluctuations and reveal the overall trend. Rolling Standard Deviation measures how much the price varies from its average over a set time, helping to understand how stable or volatile the price is. Adding these features gives My model more information, making it better at spotting patterns and improving its accuracy, just like giving a detective more clues to solve a case.

```python
data['SMA_20'] = data['Close'].rolling(window=20).mean()
data['SMA_50'] = data['Close'].rolling(window=50).mean()
data['Rolling_Std_3'] = data['Close'].rolling(window=50).std()
print(data)
```


The MACD (Moving Average Convergence Divergence) is an indicator used to understand trends and momentum in data. It calculates values like the MACD line (the difference between fast and slow averages), the Signal Line (a smoothed version of the MACD line), and the Histogram (the difference between the MACD and Signal Line). These features help my model spot changes in trends and strength of movement. Adding them provides more detailed insights into the data, improving the model’s ability to predict outcomes

```python
macd, signal, hist = talib.MACD(data['Close'], fastperiod=12, slowperiod=26, signalperiod=9)
data['MACD'] = macd
data['Signal Line'] = signal
data['Histogram'] = hist
print(data)
```

```python
plt.figure(figsize=(14, 7))
plt.plot(data.index, data['MACD'], label='MACD Line', color='blue', linewidth=2)
plt.plot(data.index, data['Signal Line'], label='Signal Line', color='red', linestyle='--', linewidth=2)
plt.bar(data.index, data['Histogram'], label='Histogram', color='green', alpha=0.5)
plt.title('MACD, Signal Line, and Histogram', fontsize=16)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.legend(loc='upper left', fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

On-Balance Volume is important in trading because it helps understand the relationship between price movement and trading volume. By measuring whether volume is flowing into or out of an asset, OBV provides clues about the strength of a trend. For example, if the price is rising but OBV is falling, it could signal a weak trend that might reverse. Unlike moving averages and MACD, which are based only on price, OBV focuses on volume as a leading indicator to predict future price movements. While moving averages and MACD show trends and momentum based on past price changes, OBV analyzes volume patterns that often signal changes before they happen. Including OBV in my machine learning model can improve accuracy by providing unique information about volume flow, helping the model identify trend reversals

```python
on_balance_volume = talib.OBV(data['Close'], data['Volume'])
data['On Balance Volume'] = on_balance_volume
print(data)
```

This feature indicates the percentage of the move in the up or down direction

```python
percentage = []
j = origional_data['Close', 'BTC-USD'][0]
for i in range(1,len(origional_data['Close','BTC-USD'])):
    percentage.append(((origional_data['Close','BTC-USD'][i] - j) / j) * 100)
    j = origional_data['Close','BTC-USD'][i]
percentage.insert(0,statistics.mean(percentage))
data['Percentage'] = percentage
data.head()
```

the diffrence feature i have added is the price diffrence between bitcoin from one day to its following day.

```python
diffrence = []
j = data['Close'][0]
for i in range(1,len(data['Close'])):
    diffrence.append(data['Close'][i] - j)
    j = data['Close'][i]
diffrence.append(sum(diffrence) / len(diffrence))
data['Diffrence'] = diffrence
data.head()
```
